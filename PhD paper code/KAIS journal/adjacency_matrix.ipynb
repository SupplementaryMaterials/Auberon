## pandas read sql server content

```python
#sql query
sql_query =  'select Han_name \
from auberon_1_top500_companies_2columns'

#using python to read the dataset from our sql server 
data = pandas.read_sql(sql_query,con)
```

```python
#storing the pandas dataframe into python list
python_list_top_500_companies = data['Han_name'].values.tolist()
```

```python
dic_500 = {}
list_500 = python_list_top_500_companies

#insert the top 500 companies into dic_500 (final will get 500 * 500 metrix through dictionary)
for i in range(len(list_500)):
    tar = list_500[i]
    dic_500[tar]={}
    for j in list_500:
        dic_500[tar][j]=0
    
```

```python
def renew_patent_matrix(dic_500,target):
    #finding the relationship matrix of International Business Machines Corporation
    sql_query = 'select Citing_Han_name,count(*) as num \
    from auberon_4_top500_companies_CitationLinkage2019_march_Han_name \
    where Cited_Han_name like \'{}\' \
    group by Citing_Han_name'.format(target)
    
    #using python to read the dataset from our sql server 
    data = pandas.read_sql(sql_query,con)
    #storing the pandas dataframe into python list
    relationship_matirx = data[['Citing_Han_name','num']].values.tolist()
    
    for i in range(len(relationship_matirx)):
        sub_tar = relationship_matirx[i][0]
        sub_tar_num = relationship_matirx[i][1]
        dic_500[target][sub_tar]=sub_tar_num
    
    result = dic_500
    return result
    
    
```

```python
#establish the patent metrix
error_list = []

for i in list_500:
    try:
        dic_500 = renew_patent_matrix(dic_500, i)
    except:
        error_list.append(i)
        pass
    
```
