# 使用多元回歸預測(兩套工具:sklearn or statsmodels)

```python
X = df[['TimeCitedFiveYearCNT','ClaimCNT']] # here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = df['uslitigation']
 
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print ('Using sklearn below:')
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)


# prediction with sklearn
New_TimeCitedFiveYearCNT = 5
New_ClaimCNT = 15
print ('Predicted US litigation: \n', regr.predict([[New_TimeCitedFiveYearCNT ,New_ClaimCNT]]))

#上下都是回歸方法，擇一使用

# with statsmodels
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print ('\n')
print ('Using statsmodels below:')
print(print_model)
```

# 解釋如何運作的:

## US litigation = (Intercept) + (TimeCitedFiveYearCNT coef)*X1 + (ClaimCNT coef)*X2
## 0.00932688 = 0.0005502499311274417 + 0.00114489 * 5 + 0.00020348 * 15
## uslitigation價格化 = 常數 + 五年引證係數*預測的X1(新的TimeCitedFiveYearCNT) + Claim數量係數*預測的X2(新的ClaimCNT)
